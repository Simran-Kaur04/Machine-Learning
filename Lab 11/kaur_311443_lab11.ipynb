```python
def remove_stopwords(file):
    new_file = []
    stop_words = set(stopwords.words('english'))
    file = re.sub('[^A-Za-z]+', ' ', file)
    file = file.lower()
    words = file.split()
    for r in words:
        if r not in stop_words and len(r) > 2:
            new_file.append(r)
    return new_file
```

```python
train = []
for f in twenty_train['data']:
    train.append(remove_stopwords(f))
```

```python
def bag_of_words(l_words):
    count = 1
    word_freq = {}
    for w in l_words:
        if w not in word_freq.keys():
            word_freq[w] = count
        else:
            word_freq[w] += count
    return word_freq
        
```

```python
train_freq = []
for i in range(len(train)):
    train_freq.append(bag_of_words(train[i]))
```

```python
train_freq
```

```python
dict_freq = {}
for d in train_freq:
    for k in d:
        if k in dict_freq:
            value = dict_freq.get(k) + d[k]
            dict_freq.update({k:value})
        else:
            dict_freq.update({k:d[k]})
```

```python
dict_freq_high = {}
for kk in sorted(dict_freq, key = dict_freq.get, reverse = True)[:100]:
        dict_freq_high[kk] = dict_freq[kk]
```

```python
def tf_idf(l_freq, d_freq):
    count = 1
    tfIdf = []
    for freq_d in l_freq:
        freq_tf = {}
        for word in freq_d.keys():
            freq_tf[word] = (freq_d[word]/len(freq_d))*np.log(len(d_freq)/d_freq[word])
        tfIdf.append(freq_tf)
    return tfIdf      
```

```python
lFreq_tfidf = tf_idf(train_freq, dict_freq)
```
